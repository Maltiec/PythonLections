##  Как правильно называть переменные

В Python нет жестких правил по наименованию переменных, таких чтобы интерпретатор сигнализировал об ошибке в коде, программист волен поступать как ему вздумается. Однако, ирония заключается в том, что с другой стороны в Python самые жесткие требования по наименованию переменных. Просто они регламентированы договоренностями между программистами. Я уже давал ссылку на замечательный документ PEP-8, в нем как раз есть раздел о нейминге.

Итак, рекомендация данного документа гласит: имя переменной пишется всеми маленькими буквами, если имя состоит из нескольких слов, то слова разделяются символом `_`. Все переменные в примерах первой лекции подчинялись этому правилу. Далее, как же придумать имя для переменной?
Допустим у нас есть значение, получаемое от датчика освещенности, и нам нужно его сохранить. В этом случае я бы выбрал:

`light_sensor_data`

или

`light_sens_data`

Почему так? Имя вашей переменной должно четко отражать ее содержимое. Помните про "откройте свой код через месяц"? Если без комментариев вы не можете сходу понять что хранится в переменной - значит имя выбрали плохое. Кроме того, не стоит указывать в имени переменной глаголы, например:

`get_light_sensor_data`

или

`getting_light_sensor_data`

Это просто вводит читающего в заблуждение, так как стиль нейминга функций в Python ровно такой же как и у переменных. Отличие только в том, что функции в своем имени как раз **должны** содержать глаголы, чтобы указывать на некое действие, выполняемое функцией. А переменные наоборот - только существительные с прилагательными. Когда начнем знакомство с функциями я приведу примеры корректного совместного нейминга для этих объектов.

Пока что вам будет непонятно, зачем решили называть переменные и функции в одинаковом стиле, но дальше, когда мы перейдем к классам и экземплярам классов - все встанет на свои места.

### Стадии разработки программы

На самом деле сам процесс "написания" программы, т.е. разработка кода - всего лишь одна из стадий процесса разработки программы. Причем эта стадия гораздо проще, чем, например, *проектирование* программы. Рабочий процесс программиста можно представить следующим образом:

- **Спроектировать программу**
  Приступая к новому проекту, программист никогда не бросается писать код, сначала необходимо сделать проект программы. Понять какую задачу или задачи программа должна выполнить, какие данные у программы на входе, а какие на выходе. Как будет строиться структура представления данных внутри самой программы, как будут взаимодействовать объекты друг с другом, и т.п. Данный этап разработки сильно влияет на конечный результат: получиться ли разработать рабочую программу или нет.

- **Написать код**
  После разработки проекта программы, программист начинает писать код на высокоуровневом языке, подчиняясь правилам синтаксиса языка. Как правило на данном этапе могут возникать синтаксические ошибки в коде, для их поиска и устранения нужен следующий этап.

- **Исправить синтаксические ошибки**
  Сразу отметим: *любая* программа содержит синтаксические ошибки, вносимые при написании кода. Даже опытный программист допускает опечатки, может использовать не тот регистр или забыть переключить раскладку клавиатуры. Синтаксическая ошибка не позволяет скомпилировать программу, таким образом пока программист не найдет все ошибки такого типа, он не сможет создать исполняемый файл. В случае запуска на интерпретаторе ситуация сложнее: программа будет запускаться и если ошибка не в самом начале кода, то даже успеет поработать, прежде чем интерпретатор "наткнется" на ошибку и не сможет продолжить дальнейшее выполнение кода. Если вы используете среду разработки, то даже для языка Python, среда укажет вам на ошибки в синтаксисе.

- **Протестировать программу**
  Когда вы привели программный код к исполнимой форме, необходимо протестировать вашу программу. Тестирование позволяет выявить второй тип ошибок - логические. Эти ошибки в целом не мешают исполнению программы, но они приводят к неверным результатам ее работы. Поиск таких ошибок на порядок сложнее, чем синтаксических.

- **Исправить логические ошибки**
  Если тестирование выявило неправильные результаты работы программы, то необходимо выполнить отладку кода - поиск и исправление логических ошибок. Иногда получается, что ошибки заложены на самом первом этапе проектирования программы, тогда придется вернуться к самому началу процесса и доработать проект. Так что разработка программ почти всегда имеет циклический характер.
  

### Разбиение сложной задачи на более простые шаги

Итак, как же подойти к самому ответственному этапу разработки программы: к проектированию? В первую очередь вам нужно понять задачу, которую должна выполнить программа. Причем не надейтесь что компьютер обладает каким-либо собственным "умом" и сможет понять задачу за вас, даже если вы используете нейронную сеть для генерации кода - непонимание вами задачи приведет к ошибочному промпту и далее, соответственно, к ошибочному коду (причем на первый взгляд он может быть даже будет работать, но не пройдет этап тестирования).

Если вы разобрались в задаче, то вы должны понимать как разбить ее на более мелкие подзадачи. Любая задача (если только не элементарная) может быть разбита на последовательность более мелких и простых задач. Далее вам нужно выстроить последовательность этих простых шагов, в результате у вас появится *алгоритм*.

### Пример проектирования программы

Например, у вас есть задача сварить кофе. Мы можем разбить данную задачу на ряд простых шагов, каждый из которых необходим для решения исходной задачи:

1. Насыпать определенное количество кофе в турку
2. Налить определенное количество воды в турку
3. Зажечь плиту
4. Поставить турку на плиту
5. Ждать появления пенки
6. Снять турку с плиты
7. Выключить плиту

Весьма линейная последовательность шагов, которая приведет нас к решению сложной задачи: получить кофе по-турецки.  Главное не нарушать последовательность этих действий и все будет хорошо - задача будет выполнена.
Обратите внимание, что я намеренно указал слова "определенное количество", там где должна быть цифра. На данный момент мы не знаем ни объем турки, ни требования к крепости кофе. Эти неопределенности нужно устранить, для чего мы теперь можем перейти к этапу определения входных и выходных данных программы.

### Определение входных и выходных данных программы

Как обычно работает программа? Чаще всего берет некие данные на входе, как-то преобразует их или обрабатывает, получает результат и выдает его на выходе. Очень похоже на решение математической задачи: у нас имеется то что дано в условии задачи, подбираем нужную последовательность действий для решения, и выдаем в конце ответ.

Для разработки рабочего проекта программы крайне важно определить что же является входными данными. В нашем примере выше с кофе, входными данными являются:

- кофе
- вода

Вы спросите, а как же все остальное? У нас же еще есть турка и плита, разве это не входные данные?

Ответ - нет. Плита и турка являются нашими инструментами для решения задачи, они должны быть внутри программы чтобы она могла выдать результат, но пользователь не должен их предоставлять для решения. А вот программист в ходе проектирования напротив должен учесть что для разработки программы ему потребуются эти объекты, без них программа работать не будет.

А что с выходными данными?

В нашем случае выходными данными является приготовленный кофе. И тут мы сразу сталкиваемся с ошибкой в нашем проекте программы: сваренный кофе находится в турке, а турка является нашим внутренним объектом, который мы не можем отдать пользователю. Следовательно, нам нужен некий контейнер для выходных данных. Хорошим вариантом контейнера является... чашка! Очевидно, что нам придется слегка доработать свой алгоритм действий, чтобы кофе в итоге оказался в контейнере, но оно того стоило. Заметьте, что мы нашли ошибку в программе, еще до того, как начали писать код.

Оператор `if` является одним из главных операторов в языке Python, его задача - выбирать действия для выполнения. ## Три базовых компонента синтаксиса Python

Те из вас кто изучал другие языки программирования сразу обратят внимание на минималистичность синтаксиса Python. Как-будто убрано большое количество визуального мусора, код выглядит значительно чище. Это и есть первый базовый компонент синтаксиса Python:

- *нет открывающих-закрывающих скобок для блоков и точек с запятыми в конце строк (конец строки = конец оператора), также не обязательны круглые скобки для условий проверок*

Помимо уборки мусора Python все же добавляет один новый элемент в синтаксис:

- *в синтаксисе Python важную роль играет двоеточие*

Все составные операторы подчинены единому шаблону:

```
Строка заголовка:
    Вложенный блок операторов
```

- *конец отступа является концом блока*

В шаблоне, приведенном выше, видно что вложенный блок смещен вправо на отступ, если же мы хотим закончить вложенный блок операторов, мы просто должны вернуться на 1 отступ влево. Отступы можно делать либо с помощью 4 пробелов, либо 1 табуляции. Смешивать эти два варианта нельзя!

Вернемся к оператору `if`. Общая форма оператора выглядит следующим образом:

```
if проверка1:     # Проверка if
    оператор1     # Связанный блок
elif проверка 2:  # Необязательный elif
    оператор2
else:             # Необязательный else
    оператор 3
```

Посмотрим на примеры применения `if`. Если у нас самый простой случай, то мы можем выкинуть из шаблона все необязательные части и выполнить только одну проверку:

````python
if 1:
    print('Истина')
````

Эта проверка всегда будет проходить, так как `1` в Python является булевским значением "истина" (в качестве эквивалента можно использовать слово `True`). Если мы хотим отработать ложный результат придется добавить `else`:

````python
if not 1:
    print('истина')
else:
    print('ложь')
````

Давайте попробуем сделать более сложный оператор `if`, содержащий все части шаблона:

````python
destination = 'Луна'
if destination == 'Марс':
    print('Место назначения - Марс!')
elif destination == 'Венера':
    print('Место назначения - Венера!')
elif destination == 'Луна':
    print('Место назначения - Луна!')
else:
    print('Никуда не летим...')
````

Что происходит в данном коде? Python выполняет операторы, вложенные внутрь первой проверки, которая дала истинный результат, или часть `else`, если все проверки показали ложные результаты. В нашем случае сработала третья проверка. И корабль отправляется к Луне.

### Булевские проверки

В языке Python существуют три булевских операции выражений.

- `X and Y` Дает истину, если истинными являются и X, и Y
- `X or Y`  Дает истину, если истинными являются либо X, либо Y
- `not X`   Дает истину, если X ложно (выражение возвращает `True` или `False`)

Как видите, в отличие от языка С, в Python булевские операции набираются как слова.

## Циклы
***

Циклические конструкции в любом языке программирования служат для многократного повторения действий, в том случае когда это необходимо. У нас уже был пример в алгоритме, когда явно нужно длительное время делать одно и то же действие - ждать. Вот здесь нам бы мог помочь цикл.

### Цикл while

Оператор `while` - самый универсальный, когда речь идет о циклах в Python. Он многократно повторяет блок операторов до тех пор, пока проверка в его заголовке оценивается как истинное значение. Если же проверка дает ложное значение даже на первом проходе - цикл не выполняется ни разу.

Как обычно приведу общую форму оператора:

```
while проверка:      # Проверка цикла
    операторы        # Тело цикла
else:                # Необязательная часть else
    операторы        # Выполняются, если не произведен выход из цикла с помощью break
```

И пара примеров применения:

````python
a = 0
b = 10
while a < b:
    print(a, end=' ') # Выводим значения переменной a в одну строку
    a = a + 1
````

В этом примере есть две особенности, на которые стоит обратить внимание. Во-первых, мы познакомились с условиями сравнений. Здесь приведено строгое неравенство a < b, поэтому a никогда не будет равно 10, цикл остановится на 9. Если бы мы записали `a <= b`, то значение 10 было бы последним.
Кроме того, вот это выражение `a = a + 1`, можно заменить более удобной и краткой конструкцией `a += 1`. Обе конструкции идентичны.

Итак, операторы сравнений:
```
X < Y   Возвращает истину если X меньше Y
X > Y   Возвращает истину если X больше Y
X <= Y  Возвращает истину если X меньше или равен Y
X >= Y  Возвращает истину если X больше или равен Y
X == Y  Возвращает истину если X равен Y
```
Посмотрим на более сложный пример цикла. У нас есть пример с оценкой топлива ракеты для выбора цели полета, теперь мы можем эмулировать сам полет:

````python
fuel_volume = 10.0  # Переменная для хранения объема топлива в баке
is_rocket_moving = True # Переменная-флаг для обозначения состояния ракеты - летит-не летит
distance_to_moon = 20 # Переменная для хранения дальности полета до Луны
while is_rocket_moving:          # Проверка условия истинности для продолжения: ракета летит или нет
    fuel_volume -= 0.01           # Уменьшаем количество топлива в баке
    distance_to_moon -= 1        # Сокращаем дистанцию до Луны
    print('Продолжаем полет')    # Выводим в консоль сообщение о продолжении полета
    if distance_to_moon == 0:    # Делаем проверку не долетели ли мы до Луны
        is_rocket_moving = False # Если да, то останавливаем ракету
        print('Добро пожаловать на Луну!')
````

В этом примере есть одна особенность, мы проверяем дистанцию до Луны *после* того как прошли очередной цикл полета, зачем именно в этом месте делать проверку? Посмотрите на блок кода ниже, здесь я для удобства экипажа вывожу оставшееся расстояние до Луны.

````python
fuel_volume = 10.0  # Переменная для хранения объема топлива в баке
is_rocket_moving = True # Переменная-флаг для обозначения состояния ракеты - летит-не летит
distance_to_moon = 20 # Переменная для хранения дальности полета до Луны
while is_rocket_moving:          # Проверка условия истинности для продолжения: ракета летит или нет
    fuel_volume -= 0.01           # Уменьшаем количество топлива в баке
    distance_to_moon -= 1        # Сокращаем дистанцию до Луны
    print('Продолжаем полет ' + 'Расстояние до Луны ' + str(distance_to_moon))    # Выводим в консоль сообщение о продолжении полета
    if distance_to_moon == 0:    # Делаем проверку не долетели ли мы до Луны
        is_rocket_moving = False # Если да, то останавливаем ракету
        print('Добро пожаловать на Луну!')
````

Как видите, мы закончили цикл полета вовремя, как раз в момент когда достигли Луны. А что если переместить проверку выше?

````python
fuel_volume = 10.0  # Переменная для хранения объема топлива в баке
is_rocket_moving = True # Переменная-флаг для обозначения состояния ракеты - летит-не летит
distance_to_moon = 20 # Переменная для хранения дальности полета до Луны
while is_rocket_moving:          # Проверка условия истинности для продолжения: ракета летит или нет
    if distance_to_moon == 0:    # Делаем проверку не долетели ли мы до Луны
        is_rocket_moving = False # Если да, то останавливаем ракету
        print('Добро пожаловать на Луну!')
    fuel_volume -= 0.01           # Уменьшаем количество топлива в баке
    distance_to_moon -= 1        # Сокращаем дистанцию до Луны
    print('Продолжаем полет ' + 'Расстояние до Луны ' + str(distance_to_moon))    # Выводим в консоль сообщение о продолжении полета
````

Мы выполнили цикл 1 лишний раз и врезались в Луну. Избежать подобной печальной участи можно с помощью оператора `break`. Он прервет цикл сразу, без возврата к исходной точке и выполнения оставшихся операторов:

````python
fuel_volume = 10.0  # Переменная для хранения объема топлива в баке
is_rocket_moving = True # Переменная-флаг для обозначения состояния ракеты - летит-не летит
distance_to_moon = 20 # Переменная для хранения дальности полета до Луны
while is_rocket_moving:          # Проверка условия истинности для продолжения: ракета летит или нет
    if distance_to_moon == 0:    # Делаем проверку не долетели ли мы до Луны
        is_rocket_moving = False # Если да, то останавливаем ракету
        print('Добро пожаловать на Луну!')
        break
    fuel_volume -= 0.1           # Уменьшаем количество топлива в баке
    distance_to_moon -= 1        # Сокращаем дистанцию до Луны
    print('Продолжаем полет ' + 'Расстояние до Луны ' + str(distance_to_moon))    # Выводим в консоль сообщение о продолжении полета
````
